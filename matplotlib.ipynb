# plt.plot Fonksiyonu

```python
# modellediğimiz ilişki- fonksiyon : f(x) = 2x
plt.figure(figsize = (10,8))  #grafiğin boyutlarını böyle değiştirebiliriz.
x = np.arange(0, 7)
y = 2 * x
plt.plot(x, y) #plot grafiği.
plt.show()
```

# plt.scatter

```python
#plt.scatter() fonksiyonu, Matplotlib kütüphanesinde nokta grafiği (scatter plot) çizmek için kullanılır. 
#Nokta grafiği, iki değişken arasındaki ilişkiyi göstermek için x ve y eksenlerinde noktaların dağılımını çizerek görselleştirir.
#Bu tür grafikler, iki değişken arasında doğru veya ters ilişkiler gibi eğilimleri ve kümelenmeleri görmeye yardımcı olabilir.

#plt.scatter() fonksiyonu, plt.plot() fonksiyonuna benzer şekilde çalışır, ancak noktaları birleştiren düz çizgiler yerine veri noktalarının dağılımını gösterir.
#İki eşit uzunluktaki liste veya dizi alır ve bunları x ve y eksenlerinde değerler olarak kullanarak bir nokta grafiği oluşturur.
```

```python
# modellediğimiz ilişki - fonksiyon = f(x) = 2x
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]

plt.figure(figsize = (6, 4))
plt.scatter(x, y, color= "magenta")
plt.plot(x, y, color= "blue")
plt.show()
```

```python
#Soru: Numpy kullanarak cos(x) ve sin(x) grafiklerini çizdirelim.
#cos(x) plt.scatter, sin(x) plt.plot
# x = np.arange(100) / 10
```

```python
x = np.arange(100) / 10
y1 = np.sin(x)
y2 = np.cos(x)
plt.scatter(x, y2, label= "cosx", color= "red")
plt.plot(x, y1, label= "sinx", color= "black")
plt.legend(loc= "upper right")
plt.show()
```

## plt.plot ve plt.scatter Parametreleri

### plt.plot 
`color` : çizginin rengini belirler.

`linestyle` : çizgi stilini belirler.

`marker` : noktalara denk gelen kesişimleri gösterir. "o" olabilir "s", "x" olabilir.

`markerfacecolor` : marker rengini değiştirir

`alpha` : saydamlık - opaklık. 1 opak, 0 saydamdır.

`markersize`:

`label` : grafik etiketi

```python
plt.plot(x, y1, color = "green", label = "Sinx",
         linestyle="--", )

plt.legend(loc="upper right")
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]

plt.figure(figsize = (6, 4))

plt.plot(x, y, color="blue", marker="s", linestyle= "--",
         markersize=7, alpha=0.7, label= "deneme", markerfacecolor= "red")
plt.legend()
plt.show()
```

## plt.scatter parametreleri
`color`: renk belirtir

`label`: grafiğin etiketi

`s`: noktaların büyüklüğü

`edgecolors`: noktaları saran renk.

`alpha`: saydamlık ve opaklık

`linewidths` : sınırdaki rengin büyüklüğü

```python
x = np.arange(100) / 10

plt.scatter(x , np.cos(x), color= "red",
           label= "Cosx", s= 40, edgecolors= "blue",
           alpha= 0.6, linewidths= 2)
plt.legend()
plt.show()
```

## Parametreler ve Özet

Parametrelerin birçoğu hem `plt.scatter()` hem de `plt.plot()` fonksiyonlarında kullanılabilir, ancak fonksiyonların kendine özgü özellikleri vardır. Örneğin, `plt.scatter()` fonksiyonu, veri noktalarının boyutunu ve renklerini değiştirmenize olanak tanırken,` plt.plot()` fonksiyonu, çizgi grafiği çizmek için daha fazla seçenek sunar.

İki fonksiyonun farklı kullanım amaçları vardır. `plt.scatter()` daha çok iki değişken arasındaki ilişkiyi ve dağılımı göstermek için kullanılırken, `plt.plot()` genellikle zaman serisi verileri veya sürekli fonksiyonlar gibi bağlantılı verileri göstermek için kullanılır.

```python
data = {
    "Yıl" : np.arange(2010, 2021),
    "Satış" : np.random.randint(100, 500, 11),
    "Masraf" : np.random.randint(50, 400, 11),
}
df = pd.DataFrame(data)
#bölge sütunu ekleyelim 
regions = ["Kuzey", "Güney", "Doğu", "Batı"]
df["Bölge"] = np.random.choice(regions,df.shape[0]) #11 tane yapmak: df.shape[0]
df
```

```python
#plt.plot ile bunu grafiğe dökelim
#yıllara göre satışları gösterelim, daha sonra yıl ve masrafı gösteren ikinci grafiği ekleyelim
```

```python
plt.plot(df["Yıl"], df["Satış"], label= "Satışlar",
        linestyle= "--", color= "blue", marker= "o",
        markerfacecolor= "red")
plt.plot(df["Yıl"], df["Masraf"], label= "Masraflar",
        linestyle= "-", color= "green", marker= "x",
        markerfacecolor= "blue")

#Başlık oluşturma, x ve y'ye isim verme

plt.title("Yıllara göre Satış ve Masraflar", fontsize = 18) #fontsize ile büyüklüğünü ayarlayabiliriz.
plt.xlabel("Yıllar", fontsize= 12)
plt.ylabel("$", fontsize= 12)

####################
plt.legend()
plt.show()
```

# plt.bar()

`plt.bar()` bir sütun grafiği çizer. Bu grafik türü, kategorilerin değerleri arasındaki ilişkiyi gösterir.



`color` : Sütunların rengini belirtir. 

`width` : Sütunların genişliğini kontrol eder

`edgecolor` : Sütunların kenarlık rengini belirler

`linewidth` : Kenarlık kalınlığını belirtir.


```python
#Soru: Aşağıda verilen veri kümesindeki Şehir ve Nüfus değerlerini kullanarak bir barplot çiziniz:
data = { 'Şehir': ['Ankara', 'İstanbul', 'İzmir', 'Bursa', 'Antalya'], 
         'Nüfus': [5503985, 15029231, 4367251, 3056120, 2444535] }
```

```python
plt.bar(data["Şehir"], data["Nüfus"], color= "grey", width= 0.85,
       edgecolor= "black" , linewidth= 1.5)
plt.title("Şehirler & Nüfus", fontsize= 15)
plt.xlabel("Şehirler")
plt.ylabel("Nüfus")
plt.show()

#le7 7 tane sıfır olduğunu gösterir (scientific notation)
```

## Histogram Mantığı

```python
sales_bins = pd.cut(df["Satış"], bins=10)
# iki fonksiyon var. pd.cut ve pd.qcut
# bin: numerik datanın grubu olarak düşünülebilir
# - (114.999, 175.0] --> bin'i ifade eder.
# pandas cut: bin arası mesafe eşit - her bin'de eşit sayıda eleman olmayabilir.
# bin aralığı: 175 - 114 = 61, 236 - 175 = 61 gibi..

# pandas qcut: bin arası mesafe eşit değil - her bin'de eşit sayıda eleman olacak şekilde ayırır.
bin_counts = sales_bins.value_counts().sort_index()

# bin'lerdeki aralığın orta değerini alır
bin_middles = [(interval.left + interval.right) / 2 for interval in bin_counts.index]

plt.bar(bin_middles,
        bin_counts.values,
        width = 30, alpha = 0.7,
        color="red")
```

# plt.hist()

`plt.hist()` fonksiyonu, verilerin dağılımını göstermek için bir histogram grafiği çizer. Histogramlar, verilerin sıklığını veya yoğunluğunu görselleştirmek için kullanılır.

#### Parametreler
`bins`: Grup sayısını belirler.

`range`: Bin'lerin alt ve üst limitini belirler.

`density`: True olursa grafik yoğunluk belirtir, false olursa frekans belirtir.

`color`: Bin rengi

`alpha`: Saydamlık opaklık

```python
df

plt.hist(df["Satış"], bins= 10,
        color= "blue", alpha= 0.5, label= "Satışların Frekansı")
plt.legend()
plt.show()
```

```python
#Soru: numpy ile random veri oluşturup plt.hist() ile frekansını görselleştirin.
#np.random.randn faydalanabilirsiniz.
#Veri sayısı 1000 olsun.
#bins = 30 olsun
#renk vb. farketmez.
```

```python
data_normal = np.random.randn(100)

```

```python
plt.hist(data_normal, bins= 30,
        color= "brown", edgecolor= "blue",
        label= "frekans", alpha= 0.8 )
plt.legend()
plt.show()
```

```python
# Min-Max değişti
plt.hist(data_normal, bins=30, color="brown", edgecolor= "blue",
         range=(-3, 3))
plt.show()
```

```python
# y ekseni değişti, yoğunluk gosteriyor, normalize değerler.
plt.hist(data_normal, bins=30, color="brown",
         edgecolor= "blue", range=(-3, 3), density = True)
plt.show()
     
```
