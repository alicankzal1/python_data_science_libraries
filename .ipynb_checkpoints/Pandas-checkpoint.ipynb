```python
#amaç : toplam hesap 10 dolardan az ise "$", 10 ile 30 arasında ise "$$", 30dan fazla ise "$$$" işareti bulunana Fiyat_Kategori adında yeni sütun oluştur.
```

## 1.yol 

```python
def fiyat_kategori(x):
    if x < 10:
        return "$"
    elif 10 <= x < 30 :
        return "$$"
    else:
        return "$$$"
```

```python
df["Fiyat_Kategori"] = df["total_bill"].apply(fiyat_kategori)
df["Fiyat_Kategori"].sample(15)
```

## 2.yol : lambda

```python
#df["Fiyat_Kategori"] = df["total_bill"].apply(lambda x : "$" if x < 10 else ("$$" if 10 <= x < 30 else "$$$"))
df.head()
```

```python
df.drop("Fiyat_Kategori", axis=1, inplace=True)
df.head()
```

## 3.yol (Vektorized)

```python
# bu yolu pek anlamadım. 
```

```python
def fiyat_kategori_vektorized(bill):
    categories = np.full_like(bill, "$", dtype="U3")
    categories[(bill >= 10) & (bill < 30)] = "$$"
    categories[bill >= 30] = "$$$"
    return categories
df["fiyat_kategori"] = fiyat_kategori_vektorized(df["total_bill"].values)
df.sample(10)
```

# .groupby() metodu

```python
# .groupby(sütun ismi) bu kullanım ile bir sütundaki değerlere göre gruplama yapılabilir. 
# Devamında grupladıktan sonra bir aggregate function'ına bakmalıyız
```

## Common Options:


```python
"""
mean(): Compute mean of groups
sum(): Compute sum of group values
size(): Compute group sizes
count(): Compute count of group
std(): Standard deviation of groups
var(): Compute variance of groups
sem(): Standard error of the mean of groups
describe(): Generates descriptive statistics
first(): Compute first of group values
last(): Compute last of group values
nth() : Take nth value, or a subset if n is a list
min(): Compute min of group values
max(): Compute max of group values

Full List at the Online Documentation: https://pandas.pydata.org/docs/reference/groupby.html

"""
```

```python
# Soru : Mpg performansını yıllara göre kıyasla.
```

```python
df["model_year"].unique()
```

```python
df.model_year.value_counts()
```

```python
df.groupby("model_year")["mpg"].mean() #yıllara göre mpg'nin ortalaması
# Yorum : 80'li yıllara gelindiğinde muhtemelen arabaların daha hızlı gitmesinden kaynaklı yakıt tüketimleri de artmıştır.
```

```python
df.groupby(by= ['model_year' , 'cylinders' ]).mean() #2 farklı sütunu da groupby() metoduyla kullanabiliriz.
```
